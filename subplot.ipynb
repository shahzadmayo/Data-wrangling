# Drwa the SubPlot

```python
#import the library

import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(0, 1 * np.pi,9 )
y = np.sin(x) # Change the funcation on your requirments

fig, ax = plt.subplots()
ax.plot(x, y)
plt.xlabel("Numbber ",color='white')
plt.ylabel("Pie Value", color='white')
plt.title("Random plot ", color='white')
plt.show()
```
